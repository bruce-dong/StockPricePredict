```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#read stock price data
df_technical = pd.read_csv("technical_data.csv")
#read financial statements data
df_fundamental = pd.read_csv("fundamental_data.csv")
```

```python
#explore financial statements data
display(df_fundamental)
display(df_fundamental.info())
df_fundamental.describe()
display(df_fundamental.isna().sum())
```

```python
#explore stock price data
display(df_technical)
display(df_technical.info())
df_technical.describe()
display(df_technical.isna().sum())
```

```python
# Explore Oppr trend of random 5 stocks
random_stocks = df_technical['Comcd'].unique()
random_stocks = np.random.choice(random_stocks, 5, replace=False)
plt.figure(figsize=(15, 8))
for stock in random_stocks:
    stock_data = df_technical[df_technical['Comcd'] == stock]
    plt.plot(stock_data['Date'], stock_data['Oppr'], label=stock)

plt.title('Oppr of 5 Random Stocks')
plt.xlabel('Date')
plt.ylabel('Oppr')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Outlier detection on financial statements data
def get_outliers(df, col_names):
    outliers = pd.DataFrame()
    for col in col_names:
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr
        c = df[col].copy()
        outliers[col] = c[(c < lower) | (c > upper)]

    return outliers
numerical_columns = df_fundamental.select_dtypes(include=[np.number]).columns
outliers = get_outliers(df_fundamental, numerical_columns)
display(outliers)
```
